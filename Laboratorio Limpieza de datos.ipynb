```python
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
import chardet
# look at the first ten thousand bytes to guess the character encoding
with open("Building_Permits.csv", 'rb') as rawdata:
    result = chardet.detect(rawdata.read(1000000))

# check what the character encoding might be
print(result)
```

```python
#Al leer el archivo con encoding ascii obtenemos un error 'ascii' codec can't decode byte 0xe2 in position 66: ordinal not in range(128)

databuild = pd.read_csv("Building_Permits.csv", encoding='ascii')

databuild.head()
```

```python
# Leemos el archivo con codificación utf-8
databuild = pd.read_csv("Building_Permits.csv", encoding='utf-8')


databuild.head()
```

```python
databuild.sample(5,random_state=2)
```

```python
#Quitamos los espacios en los nombres de las columnas
databuild.columns[databuild.columns != databuild.columns.str.extract(r'^(\w+)$')] 
```

```python
#Reemplazamos los espacios en los nombres de las columnas por "_"
databuild.columns = databuild.columns.str.replace(' ', '_')
databuild.head()
```

```python
#Verificamos otra vez las columnas que siguen teniendo caracteres raros
databuild.columns[databuild.columns != databuild.columns.str.extract(r'^(\w+)$')] 
```

```python
# Vamos a arreglar los dos nombres de columnas restantes

databuild=databuild.rename(columns = {"Neighborhoods_-_Analysis_Boundaries":"Neighborhoods_Analysis_Boundaries",
                           "Voluntary_Soft-Story_Retrofit":"Voluntary_Soft_Story_Retrofit"})
```

```python
#Comprobamos una última vez
databuild.columns[databuild.columns != databuild.columns.str.extract(r'^(\w+)$')] 
```

```python
#creamos un filtro de columnas que contengan datos de tipo fechas
filter_col = [col for col in databuild if ("Date" in col)]
filter_col
```

```python
#Verificamos que el filtro funciona
databuild[filter_col]
```

```python
# Convertimos las columnas que contienen fechas a tipo datetime 
for col in filter_col:
 databuild[col] = pd.to_datetime(databuild[col],infer_datetime_format=True )

databuild[filter_col]
```

```python
#Vemos que los tipos de datos se cambiaron correctamente
databuild.dtypes
```

```python
#Reordenamos la última columna "Record_ID" como la primera
databuild=databuild.set_index('Record_ID').reset_index()
databuild
```
