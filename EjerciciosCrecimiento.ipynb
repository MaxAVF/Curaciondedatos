```python
%matplotlib inline
import pandas as pd
import copy
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/wtageinf.csv')
```

```python
np.random.seed(0)
data.sample(6)
```

```python
print(data.describe())
np.random.seed(0)
data.sample(6)
```

```python
datamen = data[data["Sex"]==1]
datamen.head(4)
```

```python
datamen.describe()
```

```python
datawomen = data[data["Sex"]==2]
datawomen.head(4)
```

```python
datawomen.describe()
```

```python
datamen.plot(x="Agemos", y=["P3", "P5", "P10","P25", "P50", "P75","P90", "P95", "P97"], kind="line")
```

```python
datawomen.plot(x="Agemos", y=["P3", "P5", "P10","P25", "P50", "P75","P90", "P95", "P97"], kind="line")
```

```python

sns.set_style("darkgrid")
sns.set_context(rc={"font.size":13,"axes.titlesize":20,"axes.labelsize":26}) 
datamenb = datamen[datamen.columns.difference(["Sex","L","M","S"])].melt('Agemos', var_name='cols',  value_name='vals')
datawomenb = datawomen[datawomen.columns.difference(["Sex","L","M","S"])].melt('Agemos', var_name='cols',  value_name='vals')
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(20,20))
sns.pointplot(x="Agemos", y="vals", hue='cols', data=datamenb, ax=ax[0],
              hue_order=["P3", "P5", "P10","P25", "P50", "P75","P90", "P95", "P97"])
ax[0].set_title("Peso niños")
ax[0].legend( loc='upper left', 
           columnspacing=1.0, labelspacing=0.0,
           handletextpad=1.0, handlelength=2.5,frameon=True,facecolor="white", 
           fancybox=True, shadow=True,fontsize=16)
ax[0].set(xlabel='Edad', ylabel='Peso')
sns.pointplot(x="Agemos", y="vals", hue='cols', data=datawomenb, ax=ax[1],
             hue_order=["P3", "P5", "P10","P25", "P50", "P75","P90", "P95", "P97"])
ax[1].set_title("Peso niñas")
ax[1].legend( loc='upper left', 
           columnspacing=1.0, labelspacing=0.0,
           handletextpad=1.0, handlelength=2.5,frameon=True,facecolor="white", 
           fancybox=True, shadow=True,fontsize=16)
ax[1].set(xlabel='Edad', ylabel='Peso')
```

```python
data2 = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/lenageinf.csv')
```

```python
print(data2.describe())
np.random.seed(0)
data2.sample(6)
```

```python
#Es importante notar que en esta base de datos, tiene una fila donde repite los nombres de las variables. 
#Por tal motivo, pandas interpreta que todos los valores de la columnas son todos de tipo string, porque encuentra
#en una de sus filas, un string. Para lidiar con ello, podemos convertir todos los datos a numeros. 
#Y los strings, espurios quedan convertidos en Nan
#datamen2=pd.to_numeric(datamen2)
data2 = data2.convert_objects(convert_numeric=True)
data2.head(4)
```

```python
data2["Sex"].unique()
```

```python
datamen2 = data2[data2["Sex"]==1]
datamen2.head(4)
```

```python
datamen2.plot(x="Agemos", y=["P3", "P5", "P10","P25", "P50", "P75","P90", "P95", "P97"], kind="line",title="Altura niños")
```
